```python
crew=pd.read_csv("title.crew.tsv",sep="\t")
```

```python
def to_nconst_set(cell):
    if pd.isna(cell) or cell == "\\N":
        return set()
    return set(cell.split(","))

crew["directors"] = crew["directors"].apply(to_nconst_set)
crew["writers"] = crew["writers"].apply(to_nconst_set)
```

```python
crew.to_csv("crew2.csv",index=False)
```

```python
crew.head(10)
```

```python
# On prépare un dictionnaire indexé par nconst
principals_dict = principals_filtred.set_index("nconst").to_dict(orient="index")
```

```python
for index, row in crew.iterrows():
    tconst = row["tconst"]

    for role in ["directors", "writers"]:
        col = "director" if role == "directors" else "writer"

        for nconst in row[role]:
            if nconst in principals_dict:
                anciens = principals_dict[nconst].get(col, "")
                if anciens:
                    if tconst not in anciens.split(","):
                        principals_dict[nconst][col] = anciens + "," + tconst
                else:
                    principals_dict[nconst][col] = tconst
            else:
                principals_dict[nconst] = {"director": None, "writer": None}
                principals_dict[nconst][col] = tconst
```

```python
principals4 = pd.DataFrame.from_dict(principals_dict, orient="index").reset_index()
principals4 = principals4.rename(columns={"index": "nconst"})
```

```python
principals4.to_csv("principals4.csv",index=False)
```

```python
name=pd.read_csv("name.basics.tsv",sep="\t")
```

```python
name["primaryProfession"].dropna().str.split(",").explode().unique()
```

```python
name2=name.copy()
name2["primaryProfession"] = name2["primaryProfession"].fillna("")

condition_actor = name2["primaryProfession"].str.contains("actor")
condition_actress = name2["primaryProfession"].str.contains("actress")
condition_producer = name2["primaryProfession"].str.contains("producer")
condition_director = name2["primaryProfession"].str.contains("director")

condition_total = condition_actor | condition_actress | condition_producer | condition_director

name2 = name2[condition_total]
```

```python
name2.to_csv("name2.csv",index=False)
```

```python
# Remplacer les NaN par une chaîne vide pour éviter les erreurs
principals6=principals5.copy()
principals6["actor"] = principals6["actor"].fillna("")
principals6["actress"] = principals6["actress"].fillna("")

# Ajouter les valeurs d'actress dans actor (en évitant les doublons ou virgules inutiles)
principals6["actor"] = principals6["actor"] + "," + principals6["actress"]
principals6["actor"] = principals6["actor"].str.strip(",")


# Supprimer la colonne 'actress'
principals6 = principals5.drop(columns=["actress"])

# Renommer la colonne si tu veux la rendre neutre
principals6 = principals6.rename(columns={"actor": "actor_or_actress"})
```

```python
# On ne garde que les colonnes utiles pour fondre
cols_roles = ["actor_or_actress", "producer", "director"]

# Melt
df_melted = pd.melt(principals7, id_vars=["nconst"], value_vars=cols_roles,
                    var_name="role", value_name="tconst")

# Supprimer les lignes vides
df_exploded = df_melted.dropna().copy()

# Convertir en string + split
df_exploded["tconst"] = df_exploded["tconst"].astype(str).str.split(",")

# Exploser les listes
df_exploded = df_exploded.explode("tconst").reset_index(drop=True)


principals8 = df_exploded.groupby(["tconst", "role"])["nconst"]\
                     .apply(lambda x: ', '.join(x)).unstack().reset_index()
```
